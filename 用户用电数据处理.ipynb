```python
import pandas as pd, numpy as np
```

```python
df=pd.read_csv('data.csv',parse_dates=True,index_col=[0,1])
df.head(10)
```

```python
df=df.unstack().KWH
df.head(10)
```

```python
df.fillna(np.nan,inplace=True)
```

```python
for col in range(df.shape[1]):
    df.iloc[:,col][(df.iloc[:,col]-df.mean(axis=1)).abs()>df.std(axis=1)*3]=np.nan
```

```python
#df.mean其实是在每一行上取所有列的均值，而不是保留每一列的均值。也许简单的来记就是axis=0代表往跨行（down)，而axis=1代表跨列（across)，作为方法动作的副词
#使用0值表示沿着每一列或行标签\索引值向下执行方法
#使用1值表示沿着每一行或者列标签模向执行对应的方法
```

```python
df_max=df.max(axis=1)
df_min=df.min(axis=1)
df_mean=df.mean(axis=1)
df_median=df.median(axis=1)
df_sum=df.sum(axis=1)
df_var=df.var(axis=1)
df_skew=df.skew(axis=1) #偏度
df_kurt=df.kurt(axis=1) #峰度
```

```python
df_per10=pd.Series([(df.iloc[row,:]>(df_max*0.9).iloc[row]).sum() for row in range(df.shape[0])],index=range(1,201))
df_per10
```

```python
from collections import Counter
df_month_max=df_max.copy()
df_month_min=df_min.copy()
for row in range(df.shape[0]):
    df_month_max.iloc[row]=Counter(df.iloc[row,:][df.iloc[row,:]==df_max.iloc[row]].index.month).most_common(1)[0][0]
    df_month_min.iloc[row]=Counter(df.iloc[row,:][df.iloc[row,:]==df_min.iloc[row]].index.month).most_common(1)[0][0]
```
